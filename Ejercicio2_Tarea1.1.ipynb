```python
# Codigo que implementa un esquema numerico
# para determinar la aproximacion de Leibniz

import numpy as np
import matplotlib.pyplot as plt

# Funci칩n que aproxima pi utilizando la serie de Leibniz
def leibniz_pi(n):
    # F칩rmula de Leibniz para aproximar pi, sumando hasta n t칠rminos
    return 4 * sum((-1)**k / (2*k + 1) for k in range(n))

# Valor verdadero de pi
true_pi = np.pi

# Diferentes valores de N (n칰mero de t칠rminos en la serie)
N_values = [10, 100, 1000, 10000]

# Listas para almacenar los errores calculados
errors_abs = []  # Error absoluto
errors_rel = []  # Error relativo
errors_perc = []  # Error porcentual
errors_sq = []  # Error cuadr치tico

# Bucle para calcular los errores para cada valor de N
for N in N_values:
    approx_pi = leibniz_pi(N)
    error_abs = abs(true_pi - approx_pi)
    error_rel = error_abs / true_pi
    error_perc = error_rel * 100
    error_sq = np.sqrt(error_abs**2)

    # Almacenamos los errores en sus respectivas listas
    errors_abs.append(error_abs)
    errors_rel.append(error_rel)
    errors_perc.append(error_perc)
    errors_sq.append(error_sq)

    # Mostramos los resultados para cada valor de N
    print(f"N={N}: Error absoluto={error_abs}, Error relativo={error_rel}, Error porcentual={error_perc}%, Error cuadr치tico={error_sq}")

# Gr치fico de los errores
plt.figure()
plt.plot(N_values, errors_abs, label='Error absoluto', marker='o')  # Error absoluto
plt.plot(N_values, errors_rel, label='Error relativo', marker='s')  # Error relativo
plt.plot(N_values, errors_perc, label='Error porcentual', marker='^')  # Error porcentual
plt.plot(N_values, errors_sq, label='Error cuadr치tico', marker='x')  # Error cuadr치tico

# Configuraci칩n de los ejes
plt.xscale('log')  # Escala logar칤tmica en el eje x
plt.yscale('log')  # Escala logar칤tmica en el eje y
plt.xlabel('N')  # Etiqueta del eje x
plt.ylabel('Error')  # Etiqueta del eje y

plt.legend()
plt.title('Errores en la aproximaci칩n de pi')

# Mostrar el gr치fico
plt.show()
```
