```python
import numpy as np
import matplotlib.pyplot as plt

# Inicialización de variables
epsilon = 1.0
iteracion = 0
errores_absolutos = []
errores_relativos = []
errores_porcentuales = []
errores_cuadraticos = []
iteraciones = []

# Valor de referencia para errores (último valor de epsilon antes de la detención)
while 1.0 + epsilon != 1.0:
    epsilon /= 2
    iteracion += 1
    errores_absolutos.append(epsilon)
    errores_relativos.append(epsilon / 1.0)  # Tomando 1.0 como referencia
    errores_porcentuales.append((epsilon / 1.0) * 100)
    errores_cuadraticos.append(epsilon ** 2)
    iteraciones.append(iteracion)
    print(f"Iteración: {iteracion}, Precisión de máquina: {epsilon}")

# Restaurar el último valor de epsilon antes de la condición de salida
epsilon *= 2
print(f"Precisión de máquina: {epsilon}")

# Graficar los errores
plt.figure(figsize=(10, 6))
plt.plot(iteraciones, errores_absolutos, label='Error Absoluto', marker='o', linestyle='-')
plt.plot(iteraciones, errores_relativos, label='Error Relativo', marker='s', linestyle='--')
plt.plot(iteraciones, errores_porcentuales, label='Error Porcentual', marker='^', linestyle='-.')
plt.plot(iteraciones, errores_cuadraticos, label='Error Cuadrático', marker='x', linestyle=':')

plt.xlabel('Iteraciones')
plt.ylabel('Magnitud del Error')
plt.yscale('log')  # Escala logarítmica para mejor visualización
plt.title('Evolución de la Precisión de Máquina y Tipos de Error')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
