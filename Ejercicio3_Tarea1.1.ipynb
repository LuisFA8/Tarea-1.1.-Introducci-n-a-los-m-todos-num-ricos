```python
import matplotlib.pyplot as plt

# Codigo que implementa el calculo de errores en operaciones numericas

def calcular_errores(x, y, valor_real):
    diferencia = x - y
    error_abs = abs(valor_real - diferencia)
    error_rel = error_abs / abs(valor_real)
    error_pct = error_rel * 100
    error_cuadratico = diferencia ** 2  # Error cuadrático

    print(f"Diferencia: {diferencia}")
    print(f"Error absoluto: {error_abs}")
    print(f"Error relativo: {error_rel}")
    print(f"Error porcentual: {error_pct}%")
    print(f"Error cuadrático: {error_cuadratico}")

    return error_abs, error_rel, error_pct, error_cuadratico

valores = [(1.0000001, 1.0000000, 0.0000001), (1.000000000000001, 1.000000000000000, 0.000000000000001)]

errores_abs = []
errores_rel = []
errores_pct = []
errores_cuadraticos = []

for x, y, real in valores:
    print(f"\nPara x={x}, y={y}:")
    error_abs, error_rel, error_pct, error_cuadratico = calcular_errores(x, y, real)

    errores_abs.append(error_abs)
    errores_rel.append(error_rel)
    errores_pct.append(error_pct)
    errores_cuadraticos.append(error_cuadratico)

# Gráficas de barras
fig, axs = plt.subplots(2, 2, figsize=(14, 12))

# Gráfico de barras de error absoluto
axs[0, 0].bar(range(len(errores_abs)), errores_abs, label='Error absoluto', alpha=0.7, color='r', width=0.3)
axs[0, 0].set_title("Error Absoluto")
axs[0, 0].set_xlabel("Índice")
axs[0, 0].set_ylabel("Valor")
axs[0, 0].legend()

# Gráfico de barras de error relativo
axs[0, 1].bar(range(len(errores_rel)), errores_rel, label='Error relativo', alpha=0.7, color='b', width=0.3)
axs[0, 1].set_title("Error Relativo")
axs[0, 1].set_xlabel("Índice")
axs[0, 1].set_ylabel("Valor")
axs[0, 1].legend()

# Gráfico de barras de error porcentual
axs[1, 0].bar(range(len(errores_pct)), errores_pct, label='Error porcentual', alpha=0.7, color='orange', width=0.3)
axs[1, 0].set_title("Error Porcentual")
axs[1, 0].set_xlabel("Índice")
axs[1, 0].set_ylabel("Valor")
axs[1, 0].legend()

# Gráfico de barras de error cuadrático
axs[1, 1].bar(range(len(errores_cuadraticos)), errores_cuadraticos, label='Error cuadrático', alpha=0.7, color='g', width=0.3)
axs[1, 1].set_title("Error Cuadrático")
axs[1, 1].set_xlabel("Índice")
axs[1, 1].set_ylabel("Valor")
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```
